```python
import numpy as np
import matplotlib.pyplot as plt

def plotfunc(a):
    x1 = -5.0
    x2 = 5.0
    dx = 0.1
    
    xAxis = np.arange(x1, x2, dx)
    yValues = np.tanh(a * xAxis)
    
    plt.plot(xAxis, yValues, label=f'a = {a}')  # Adding label to the plot

def main_bc():
    plotfunc(0.5)
    plotfunc(1.3)
    plotfunc(2.2)
    
    plt.xlabel("X - axis")
    plt.ylabel("Y - axis")
    plt.legend()  # Adding legend to the plot
    plt.show()
    
    

main_bc()
```
