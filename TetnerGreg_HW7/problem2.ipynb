```python
import numpy as np
import matplotlib.pyplot as plt


def trwave(x):
    """
    Calculates the triangle-wave function.

    PARAMETERS:
        x -- input;

    RETURNS:
        Triangle wave function.
          triangle_wave = np.piecewise(x, [x < 0, (x >= 0) & (x <= np.pi)], [np.pi/2 + x, np.pi/2 - x])
    return triangle_wave
    """
  
    result = np.empty_like(x)
    for i, val in enumerate(x):
        if val >= -np.pi and val < 0:
            result[i] = np.pi/2 + val
        elif val >= 0 and val <= np.pi:
            result[i] = np.pi/2 - val
        else:
            result[i] = 0
    return result

def trwave_partial_sum(x, m):
    """
    Calculates the triangle-wave function using m Fourier terms.

    PARAMETERS:
        x -- input;
        m -- number of Fourier terms

    RETURNS:
        Triangle wave function.
    """
    triangle_wave = np.zeros_like(x)
    for n in range(1, m + 1):
        triangle_wave += ((2-2*(-1**n))/(np.pi*(n**2)))*np.cos(n*x)
    return triangle_wave


def square_wave(x):
    """
    Calculates the square-wave function.
    
    PARAMETERS:
        x -- input;

             
    RETURNS:
        Square wave function.
    """
    #sqwave = ...
    
    return np.piecewise(x, [x < -np.pi/2, (x >= -np.pi/2) & (x <= np.pi/2), x > np.pi/2], [-1, 1, -1])


def square_wave_partial_sum(x,m):
    """
    Calculates the square-wave function using m Fourier terms .
    
    PARAMETERS:
        x -- input;
        m -- number of Fourier terms

             
    RETURNS:
        Square wave function.
    """
    #sqwavePartialSum = ...
    
    square_wave = np.zeros_like(x)
    
    for n in range(1, m + 1):
        square_wave += (4 * np.sin(n * np.pi / 2) / (np.pi * n)) * np.cos(n * x)
    return square_wave
    


def main():
    x = np.linspace(-np.pi, np.pi, 1000)
    
    # Plot both the square wave and the triangle wave
    plt.figure(figsize=(10, 6))
    
    # Plot the square wave
    plt.subplot(2, 1, 1)
    plt.plot(x, square_wave(x), label='Square Wave')
    for m in [1, 5, 10, 50, 100]:
        plt.plot(x, square_wave_partial_sum(x, m), label=f'{m} Terms')
    plt.legend()
    plt.title('Square Wave and Fourier Series Approximation')
    
    # Plot the triangle wave
    plt.subplot(2, 1, 2)
    plt.plot(x, trwave(x), label='Triangle Wave')
    for m in [1, 5, 10, 50, 100]:
        plt.plot(x, trwave_partial_sum(x, m), label=f'{m} Terms')
    plt.legend()
    plt.title('Triangle Wave and Fourier Series Approximation')
    
    plt.tight_layout()
    plt.show()
    
    
    """
        # ??? Write your plotting commands here. ???
    fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(15, 8))

    for i, num_terms in enumerate(num_terms_list):
        ax = axes[i // 3, i % 3]
        ax.plot(x, fourier_series(x, num_terms), label=f'{num_terms} Terms')
        ax.plot(x, triangle_wave, 'k--', label='Triangle Wave')
        ax.set_title(f'{num_terms} Terms')
        ax.legend()

    plt.tight_layout()
    plt.show()
    
    """

    
    
    

main()  # Don't delete the call to main() by mistake!


# Part c:
# You can describe your findings here.

"""

As we add more and more terms, he fourier 
series and the original function becomes more and more similar. However,
for the square wave, there are is a much more pronounced Gibbs phenomenon. I think
this is becuase it has more sharp discontinuities between -1 to 1. 



"""
```
