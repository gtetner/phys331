```python
import numpy as np
import matplotlib.pyplot as plt

def sdft(x, inverse=False):
    """
    Calculates the forward or backward discrete Fourier transform of a function defined on an array
    
    PARAMETERS:
        array -- input;
        
        inverse -- forward/backward;

        See the assignment sheet for additional details about these
        function parameters.
             
    RETURNS:
        sdft -- output;
    """
    
    N = len(x)
    n = np.arange(N)
    k = n.reshape((N, 1))
    if inverse:
        M = np.exp(2j * np.pi * k * n / N)
    else:
        M = np.exp(-2j * np.pi * k * n / N)
    return np.dot(M, x)



def main():
    # Test
    N = 64
    x = np.linspace(-np.pi, np.pi, N)
    y = np.sin(x)  

    # Forward DFT
    yf = sdft(y)

    # Inverse DFT
    y_inv = sdft(yf, inverse=True)


    # original function
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x, y)
    plt.title('Original Function')

    # transformed function (real and imaginary parts)
    plt.subplot(1, 2, 2)
    plt.plot(x, yf.real, label='Real')
    plt.plot(x, yf.imag, label='Imaginary')
    plt.legend()
    plt.title('Transformed Function')
    plt.show()

    # Test the function
    N = 128
    x = np.linspace(-np.pi, np.pi, N)
    y = np.cos(x)  # Original function

    # Perform DFT and inverse DFT
    yf = sdft(y)
    yfi = sdft(yf, inverse=True)

    # Plot original function
    plt.figure(figsize=(12, 6))
    plt.subplot(1, 2, 1)
    plt.plot(x, y, label='Original')
    plt.title('Original Function')
    plt.legend()

    # Plot transformed function
    plt.subplot(1, 2, 2)
    plt.plot(x, yf.real, label='Real')
    plt.plot(x, yf.imag, label='Imaginary')
    plt.title('Transformed Function')
    plt.legend()

    plt.show()

main()

```
